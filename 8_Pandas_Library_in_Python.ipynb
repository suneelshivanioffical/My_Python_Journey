> ### Bar Plot

```python
y = [98,67,88,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]

plt.bar(x,y)
plt.xlabel("Part of Harry Potter")
plt.ylabel("Poplutatity")
plt.title("Poplutatity of different part of Harry Potter")
plt.show()
```

```python
y = [98,67,88,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]

plt.bar(x,y)
plt.xlabel("Part of Harry Potter", fontsize = 13)
plt.ylabel("Poplurtatity",fontsize = 13)
plt.title("Poplurtatity of different part of Harry Potter",fontsize = 14)
plt.show()
```

```python
y = [98,67,88,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]

plt.bar(x,y, color = "blue")
plt.xlabel("Part of Harry Potter", fontsize = 13)
plt.ylabel("Poplurtatity",fontsize = 13)
plt.title("Poplurtatity of different part of Harry Potter",fontsize = 14)
plt.show()
```

```python
y = [98,67,88,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]
color = ["red","green","yellow","blue","orange"]
plt.bar(x,y, color = color)
plt.xlabel("Part of Harry Potter", fontsize = 13)
plt.ylabel("Poplurtatity",fontsize = 13)
plt.title("Poplurtatity of different part of Harry Potter",fontsize = 14)
plt.show()
```

> ### Stack Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,20,35,60,80]
NOP2 = [10,60,30,75,40,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3)
plt.show()
```
